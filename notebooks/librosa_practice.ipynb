# Reading Audio 
Use librosa to load audio 

```python
import librosa
```

```python
ls ../audio  #show the current sample audio files
```

```python
audio_path="../audio/TechnoDrumMIX.wav"
y, sr = librosa.load(audio_path)
```

```python
print(y)
print(sr)
```

# Visualizing Audio 
we could use matplotlib to draw the waveform of the audio files

```python
%matplotlib inline 
# Above line could let IPython embedded the pylot 
import matplotlib.pyplot as plt 
import librosa.display
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(y, sr=sr)
```

# STFT

STFT(Short Time Fourier Transform), it computes a signal in the time-frequency domain by computing the DFT(Discrete Fourier Transform) over short overlapping windows.

```python
import numpy as np

S = np.abs(librosa.stft(y))
```

```python
S
```

# Display the spectrogram 

```python
fig, ax = plt.subplots()
img= librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log',x_axis='time',ax=ax)
ax.set_title("Power  Spectrogram")
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
